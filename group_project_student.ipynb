```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import plotly.express as px


stu = pd.read_csv('dataset.csv')

X = stu.drop('Target', axis=1)
y = stu['Target']

scaler = StandardScaler()
X_std = scaler.fit_transform(X)

pca = PCA()
X_pca = pca.fit_transform(X_std)

explained_variance = pca.explained_variance_ratio_
cumulative_explained_variance = explained_variance.cumsum()

plt.plot(range(1, len(cumulative_explained_variance) + 1), cumulative_explained_variance, marker='o')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.show()

n_components = 34

pca = PCA(n_components=n_components)
X_pca = pca.fit_transform(X_std)

columns = [f'PC{i}' for i in range(1, n_components + 1)]
stu_pca = pd.DataFrame(data=X_pca, columns=columns)
stu_pca['Target'] = y
```

## *Label Encoding at Target column, Split data into train and test model*

```python
#import libraries  accuracy and test-train slipt:

import sklearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

```python
# This command is used to Split dataset into X and y:

X = stu.drop(columns= ['Target'])
y = stu[['Target']]
```

```python
#This command is used to show the order pair of X and y:

print("X shape is: ",X.shape)
print("y shape is: ",y.shape)
```

```python
#This command split given dataset into test and train:

X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size = 0.20)
```

```python
#This command shows the order pair of test and train

print("shape of X_train: ", X_train.shape)
print("shape of Y_train: ", Y_train.shape)
print("shape of X_test: ", X_test.shape)
print("shape of Y_test: ", Y_test.shape)
```

## *Classification Model --> "Logistic Regression"*

```python
#Import Logistic Regression libraries:

from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
# Fit the Model:

model.fit(X_train, Y_train)
```

```python
#Calculate accuracy score:

y_pred = model.predict(X_test)
score = accuracy_score(Y_test, y_pred)
accuracy = score*100
print(accuracy)
```

```python
# This command is used to draw confussion matrix:

from sklearn.metrics import confusion_matrix
y_pred = model.predict(X_test)
print("Confusion Matrix")
cm_lr = confusion_matrix(Y_test, y_pred)
print(cm_lr)
```

```python
# This command is used to draw a heatmap of confussion matrix:

plt.figure(figsize = (6, 4))
sns.heatmap(cm_lr, annot = True, fmt = 'd', cmap = 'Blues', cbar = False, annot_kws = {'size' : 14})
plt.xlabel('Predicted Labels', fontsize = 14)
plt.ylabel('True Labels', fontsize = 14)
plt.title('Confusion Matrix', fontsize = 16)
plt.show()
```

```python
#This command is used to verify the model based on given dataset:

X_test1 = X_test.values
y_test1 = Y_test.values
y_pred = str(model.predict([X_test1[16]]))
print(type(y_pred))

if y_pred == "[0]":
  print("Dropout")
elif y_pred == "[1]":
  print("Enrolled")
elif y_pred == "[2]":
  print("Graduate")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Fit the Model:
model.fit(X_train, Y_train)

# Calculate accuracy score:
y_pred = model.predict(X_test)
accuracy = accuracy_score(Y_test, y_pred)
print("Accuracy:", accuracy)

# Calculate precision:
precision = precision_score(Y_test, y_pred, average='weighted')
print("Precision:", precision)

# Calculate recall:
recall = recall_score(Y_test, y_pred, average='weighted')
print("Recall:", recall)

# Calculate F1-score:
f1 = f1_score(Y_test, y_pred, average='weighted')
print("F1-score:", f1)
```

## *Classification Model --> "Support Vector Machine (SVM)"*

```python
#Import support vector machine library and develop the model:

from sklearn.svm import SVC
sss = SVC()
```

```python
#Fit model:
sss.fit(X_train, Y_train)
```

```python
#This command is used to calculate accuracy of the model:

y_pred = sss.predict(X_test)
score = accuracy_score(Y_test, y_pred)
accuracy = score*100
print(accuracy)
```

```python
# This command is used to draw a heatmap of confussion matrix:

plt.figure(figsize = (6, 4))
sns.heatmap(cm_sss, annot = True, fmt = 'd', cmap = 'Blues', cbar = False, annot_kws = {'size' : 14})
plt.xlabel('Predicted Labels', fontsize = 14)
plt.ylabel('True Labels', fontsize = 14)
plt.title('Confusion Matrix', fontsize = 16)
plt.show()
```

```python
# SVM
sss.fit(X_train, Y_train)
y_pred_svm = sss.predict(X_test)
accuracy_svm = accuracy_score(Y_test, y_pred_svm)
precision_svm = precision_score(Y_test, y_pred_svm, average='weighted')
recall_svm = recall_score(Y_test, y_pred_svm, average='weighted')
f1_svm = f1_score(Y_test, y_pred_svm, average='weighted')

# Decision Tree
dtc.fit(X_train, Y_train)
y_pred_decision_tree = dtc.predict(X_test)
accuracy_decision_tree = accuracy_score(Y_test, y_pred_decision_tree)
precision_decision_tree = precision_score(Y_test, y_pred_decision_tree, average='weighted')
recall_decision_tree = recall_score(Y_test, y_pred_decision_tree, average='weighted')
f1_decision_tree = f1_score(Y_test, y_pred_decision_tree, average='weighted')

# Logistic Regression
model.fit(X_train, Y_train)
y_pred_logistic_regression = model.predict(X_test)
accuracy_logistic_regression = accuracy_score(Y_test, y_pred_logistic_regression)
precision_logistic_regression = precision_score(Y_test, y_pred_logistic_regression, average='weighted')
recall_logistic_regression = recall_score(Y_test, y_pred_logistic_regression, average='weighted')
f1_logistic_regression = f1_score(Y_test, y_pred_logistic_regression, average='weighted')

print("SVM Metrics:")
print("Accuracy:", accuracy_svm)
print("Precision:", precision_svm)
print("Recall:", recall_svm)
print("F1-score:", f1_svm)
print()

print("Decision Tree Metrics:")
print("Accuracy:", accuracy_decision_tree)
print("Precision:", precision_decision_tree)
print("Recall:", recall_decision_tree)
print("F1-score:", f1_decision_tree)
print()

print("Logistic Regression Metrics:")
print("Accuracy:", accuracy_logistic_regression)
print("Precision:", precision_logistic_regression)
print("Recall:", recall_logistic_regression)
print("F1-score:", f1_logistic_regression)
```

The three models (SVM, Decision Tree, Logistic Regression) show relatively similar performances in terms of accuracy, precision, recall, and F1-score. Logistic Regression seems to have a slight edge in terms of accuracy and recall.

```python
import pandas as pd
import matplotlib.pyplot as plt
 
stu['Tuition Group'] = stu['Tuition fees up to date'].map({0: 'Not Up to Date', 1: 'Up to Date'})
 
nombre_etudiants_par_groupe = stu.groupby(['Target', 'Tuition Group']).size().unstack()
 
fig, ax = plt.subplots()
bar_plot = nombre_etudiants_par_groupe.plot(kind='bar', stacked=True, ax=ax, color=['navy', 'skyblue'])
 
ax.set_xlabel('Category')
ax.set_ylabel('Number of students')
ax.set_title('Number of students by category and status of tuition fees')
 
ax.set_xticklabels(['Dropout', 'Enrolled', 'Graduated'], rotation=0)
 
ax.legend(title='Tuition Fees Up to Date', loc='upper right')
 
for p in bar_plot.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.text(x + width/2, y + height/2, f'{int(height)}', ha='center', va='center', fontsize=8, color='white')
 
plt.show()
```
